```python
import pandas as pd  
# Mengimpor pandas untuk membaca dan mengolah data

import matplotlib.pyplot as plt  
# Mengimpor matplotlib untuk membuat grafik

myDataset1 = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# Membaca file Excel dan mengambil data dari sheet "Supp_traces"

x = myDataset1.Zr  
# Mengambil kolom Zr sebagai sumbu X
y = myDataset1.Th  
# Mengambil kolom Th sebagai sumbu Y

loc_parameters = ['upper right','upper left','lower left','lower right','center','center left']  
# List posisi legenda yang akan diuji

fig = plt.figure(figsize=(8,4))  
# Membuat figure dengan ukuran lebar 8 dan tinggi 4

for i in range(len(loc_parameters)):  
    # Perulangan untuk setiap posisi legenda
    ax = fig.add_subplot(2,3,i+1)  
    # Membuat subplot (2 baris, 3 kolom, posisi ke-i+1)
    
    ax.scatter(x, y, marker='s', color='#c7ddf4', edgecolor='#000000', label="loc = " + loc_parameters[i])  
    # Scatter plot dengan marker kotak (s), warna biru muda, dan outline hitam
    
    ax.set_xlabel("Zr [ppm]")  
    ax.set_ylabel("Th [ppm]")  
    
    ax.legend(loc=loc_parameters[i])  
    # Menampilkan legenda dengan posisi sesuai list

fig.tight_layout()  
# Mengatur tata letak agar subplot tidak bertumpuk
```

```python
import pandas as pd  
# Mengimpor pandas untuk membaca dan mengolah data

import matplotlib.pyplot as plt  
# Mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel(
    'Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# Membaca file Excel dan mengambil data dari sheet "Supp_traces"

my_dataset1 = my_dataset[my_dataset.Epoch == 'one']  
# Memfilter data untuk Epoch = 'one'

my_dataset2 = my_dataset[my_dataset.Epoch == 'two']  
# Memfilter data untuk Epoch = 'two'

fig = plt.figure()  
# Membuat figure

ax1 = fig.add_subplot(2, 1, 1)  
# Subplot pertama (2 baris, 1 kolom, posisi 1)

ax1.scatter(my_dataset1.Zr, my_dataset1.Th, marker='s', color='#c7ddf4', edgecolor='#000000', label="First Epoch")  
# Scatter plot untuk Epoch pertama (marker kotak, warna biru muda)

ax1.scatter(my_dataset2.Zr, my_dataset2.Th,  marker='o', color='#ff464a', edgecolor='#000000', label="Second Epoch")  
# Scatter plot untuk Epoch kedua (marker lingkaran, warna merah)

ax1.set_xlabel("Zr [ppm]")  
ax1.set_ylabel("Th [ppm]")  

ax1.legend(loc='upper left', framealpha=1, frameon=True, title="Age < 15 ky", title_fontsize=10)  
# Menampilkan legenda:
# loc → posisi legenda
# frameon=True → menampilkan kotak legenda
# framealpha=1 → transparansi kotak (1 = solid)
# title → judul legenda
# title_fontsize → ukuran font judul legenda

ax2 = fig.add_subplot(2, 1, 2)  
# Subplot kedua (posisi bawah)

ax2.scatter(my_dataset1.Zr, my_dataset1.Th, marker='s', color='#c7ddf4', edgecolor='#000000', label="First Epoch")  
# Scatter plot ulang untuk Epoch pertama

ax2.scatter(my_dataset2.Zr, my_dataset2.Th,  marker='o', color='#ff464a', edgecolor='#000000', label="Second Epoch")  
# Scatter plot ulang untuk Epoch kedua

ax2.set_xlabel("Zr [ppm]")  
ax2.set_ylabel("Th [ppm]")  

ax2.legend(frameon=False, loc='lower right', ncol=2, title="Age < 15 ky", title_fontsize=10)  
# Legenda tanpa kotak (frameon=False)
# loc → posisi legenda di kanan bawah
# ncol=2 → legenda ditampilkan dalam 2 kolom

fig.tight_layout()  
# Mengatur tata letak agar tidak bertumpuk
```

```python
# Explicit positive and negative reporting
a = +5.34352
b = -6.3421245
print('-------------------------------------------------')
print("The plus symbol is not reported: {:.2f} | {:.2f}".format(+5.34352, -6.3421245))
print("The plus symbol is reported: {:+.2f} | {:+.2f}".format(a, b))
print('-------------------------------------------------')

# Reporting as percent
c = 0.1558
print('-------------------------------------------------')
print("Reporting as percent: {:.1%}".format(c))
print('-------------------------------------------------')

# Scientific notation
d = 6580000000000
print('-------------------------------------------------')
print("Scientific notation using e: {:.1e}".format(d))
print("Scientific notation using E: {:.1E}".format(d))
print('-------------------------------------------------')

'''Results
-------------------------------------------------
The plus symbol is not reported: 5.34 | -6.34
The plus symbol is reported: +5.34 | -6.34
-------------------------------------------------
-------------------------------------------------
Reporting as percent: 15.6%
-------------------------------------------------
-------------------------------------------------
Scientific notation using e: 6.6e+12
Scientific notation using E: 6.6E+12
-------------------------------------------------
'''




```

```python
import pandas as pd  
# Mengimpor pandas untuk membaca dan mengolah data

import matplotlib.pyplot as plt  
# Mengimpor matplotlib untuk membuat grafik

import numpy as np  
# Mengimpor numpy untuk operasi numerik (array, linspace, dll)


def my_line(x, m, q):  
    y = m * x + q  
    # Membuat persamaan garis lurus (y = mx + q)
    return y  
    # Mengembalikan nilai y


my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')  
# Membaca file Excel (sheet Supp_majors) menggunakan engine openpyxl

my_dataset1 = my_dataset[my_dataset.Epoch == 'one']  
# Memfilter data untuk Epoch pertama

my_dataset2 = my_dataset[my_dataset.Epoch == 'two']  
# Memfilter data untuk Epoch kedua

x = np.linspace(52.5, 62, 100)  
# Membuat 100 titik nilai x dari 52.5 sampai 62

y = my_line(x, m=0.3, q=-10.3)  
# Menghitung nilai y menggunakan fungsi garis lurus


fig, ax = plt.subplots()  
# Membuat figure dan axes

ax.scatter(my_dataset1.SiO2, my_dataset1.K2O, marker='s', color='#c7ddf4', edgecolor='#000000', label=r'$1^{st}$ Epoch')  
# Scatter plot untuk Epoch pertama (menggunakan format LaTeX untuk label)

ax.scatter(my_dataset2.SiO2, my_dataset2.K2O, marker='s', color='#ff464a', edgecolor='#000000', label=r'$2^{nd}$ Epoch')  
# Scatter plot untuk Epoch kedua

ax.plot(x, y, color='#342a77')  
# Plot garis hasil fungsi linear


ax.annotate(r'What is the 1$\sigma$ for this point?', xy=(47.6, 6.6), xytext=(47, 8.8), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))  
# Menambahkan anotasi teks dengan panah ke titik tertentu

ax.text(52.4, 5.6, r'$ Na_2O = 0.3 \cdot SiO_2 -10.3$', dict(size=10, rotation=33))  
# Menambahkan teks persamaan garis pada grafik (dengan rotasi)

ax.text(53.5, 5.1, r'$ \mu_{SiO_2} = \frac {a_{1}+a_{2}+\cdots +a_{n}}{n}$ = ' + '{:.1f} [wt.%]'.format(57.721),  dict(size=11.5))  
# Menambahkan teks rata-rata (mean) SiO2 dengan format LaTeX + nilai numerik


ax.set_xlabel(r'SiO$_2$ [wt%]')  
# Label sumbu X dengan format kimia (subscript)

ax.set_ylabel(r'K$_2$O [wt%]')  
# Label sumbu Y

ax.legend()  
# Menampilkan legenda
```

```python
import matplotlib.pyplot as plt  
# Mengimpor matplotlib untuk membuat grafik

import numpy as np  
# Mengimpor numpy untuk operasi numerik


def ec(f, d, c0):  
    cl = c0 / (d * (1-f) + f)  
    # Menghitung konsentrasi berdasarkan rumus
    return cl  
    # Mengembalikan hasil


my_f = np.linspace(0.1, 1, 10)  
# Membuat 10 nilai dari 0.1 sampai 1

my_c1 = ec(f=my_f, d=0.1, c0=100)  
# Menghitung nilai C untuk setiap nilai F

colors = ['#ff9494', '#cbeaa2', '#d1a396', '#828fc3', '#95b2e5', '#e9b8f4', '#f4b8e5', '#b8f4f2', '#c5f4b8', '#f9ca78']  
# List warna untuk setiap titik data

fig = plt.figure()  
# Membuat figure

ax1 = fig.add_subplot(2, 2, 1)  
# Subplot pertama
ax1.plot(my_f, my_c1, marker='o', linestyle='-', markersize=5)  
# Grafik garis dengan marker titik
ax1.set_xlabel('F')
ax1.set_ylabel('C [ppm]')

ax2 = fig.add_subplot(2, 2, 2)  
# Subplot kedua
ax2.scatter(my_f, my_c1, marker='o', s=my_f*150)  
# Scatter plot dengan ukuran titik bervariasi (berdasarkan nilai F)
ax2.set_xlabel('F')
ax2.set_ylabel('C [ppm]')

ax3 = fig.add_subplot(2, 2, 3)  
# Subplot ketiga
ax3.scatter(my_f, my_c1, marker='o', c=colors, s=my_f*150)  
# Scatter plot dengan warna berbeda untuk tiap titik
ax3.set_xlabel('F')
ax3.set_ylabel('C [ppm]')

ax4 = fig.add_subplot(2, 2, 4)  
# Subplot keempat
ax4.plot(my_f, my_c1, marker='', linestyle='-', zorder=0)  
# Plot garis di belakang (zorder=0 → layer bawah)

ax4.scatter(my_f, my_c1, marker='o', c=colors, s=my_f*150, zorder=1)  
# Scatter plot di atas garis (zorder=1 → layer atas)

ax4.set_xlabel('F')
ax4.set_ylabel('C [ppm]')

fig.tight_layout()  
# Mengatur tata letak agar tidak bertumpuk
```

```python
import pandas as pd  
# Mengimpor pandas untuk membaca dan mengolah data

import matplotlib.pyplot as plt  
# Mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel(
    'Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# Membaca file Excel dan mengambil data dari sheet "Supp_traces"

epochs = ['one', 'two', 'three', 'three-b']  
# List kategori Epoch

colors = ['#c8b4ba', '#f3ddb3', '#c1cd97', '#e18d96']  
# List warna untuk tiap Epoch

markers = ['o', 's', 'd', 'v']  
# List bentuk marker untuk tiap Epoch

fig, ax = plt.subplots()  
# Membuat figure dan axes

for (epoch, color, marker) in zip(epochs, colors, markers):  
    # Loop bersamaan untuk epoch, warna, dan marker
    my_data = my_dataset[(my_dataset.Epoch == epoch)]  
    # Memfilter data berdasarkan epoch
    
    ax.scatter(my_data.Zr, my_data.Th, marker=marker, s=50, c=color, edgecolor='0', label="Epoch " + epoch)  
    # Scatter plot dengan marker, warna, dan label berbeda untuk tiap epoch

ax.set_xlabel("Zr [ppm]")  
# Label sumbu X

ax.set_ylabel("Th [ppm]")  
# Label sumbu Y

ax.legend(title="Phlegraean Fields \n Age < 15 ky")  
# Menampilkan legenda dengan judul (menggunakan newline \n)
```
