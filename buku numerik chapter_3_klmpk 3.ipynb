```python
# Menyelesaikan sistem persamaan linear dua variabel
eq1 = x + 2 * y - 1
eq2 = x - y + 1
sympy.solve([eq1, eq2], [x, y], dict=True)
```

```python
# Menyelesaikan sistem persamaan non-linear
eq1 = x**2 - y
eq2 = y**2 - x
sols = sympy.solve([eq1, eq2], [x, y], dict=True)
sols
```

```python
# Mengecek apakah setiap solusi benar-benar memenuhi kedua persamaan
[eq1.subs(sol).simplify() == 0 and eq2.subs(sol).simplify() == 0 
for sol in sols]
```
