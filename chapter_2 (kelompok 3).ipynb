```python
# menampilkan beberapa urutan angka dengan range yang berbeda

print('urutan dari 0 sampai 3')
for i in range(4):
    print(i)
    
print('----------------------')    
print('urutan dari 4 sampai 9')
for i in range(4, 10):
    print(i)

print('----------------------')    
print('urutan dari 1 sampai 9 dengan langkah 2')
for i in range(1 , 10 , 2):
    print(i)
```

```python
# menampilkan angka mulai dari 7 sampai kurang dari 10 dengan perulangan while

angka = 7  

while angka < 10:  
    print(angka)
    angka = angka + 1 
```
